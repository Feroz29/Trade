# Calculating Financial Metrics

```python
# Function to calculate financial metrics
def calculate_metrics(trade_history):
    trades = pd.DataFrame(trade_history)
    
    # Calculate ROI
    total_investment = trades['quantity'].sum()
    total_profit = trades['realizedProfit'].sum()
    roi = total_profit / total_investment if total_investment else 0
    
    # Calculate PnL
    pnl = total_profit
    
    # Calculate Sharpe Ratio
    returns = trades['realizedProfit'] / trades['quantity']
    sharpe_ratio = returns.mean() / returns.std() if returns.std() else 0
    
    # Calculate Maximum Drawdown (MDD)
    cumulative_returns = (1 + returns).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (cumulative_returns - peak) / peak
    mdd = drawdown.min()
    
    # Calculate Win Rate, Win Positions, Total Positions
    win_positions = (trades['realizedProfit'] > 0).sum()
    total_positions = len(trades)
    win_rate = win_positions / total_positions if total_positions else 0
    
    return roi, pnl, sharpe_ratio, mdd, win_rate, win_positions, total_positions

# Apply the function to each account
metrics = data['Trade_History'].apply(calculate_metrics)
metrics_df = pd.DataFrame(metrics.tolist(), columns=['ROI', 'PnL', 'SharpeRatio', 'MDD', 'WinRate', 'WinPositions', 'TotalPositions'])
data = pd.concat([data, metrics_df], axis=1)
```
